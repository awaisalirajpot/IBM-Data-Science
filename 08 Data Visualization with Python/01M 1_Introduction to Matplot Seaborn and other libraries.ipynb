## **Types of Plots**
**Line plot:**
- Display trends over time
- Compare data sets with a continuous independent variable
- Illustrate cause-and-effect relationships.
- Visualize Continuous data
<br>

**Bar Plot**
- Displays data using rectangular bars
- The height or length of the bars represents the magnitude of the data
- Displays bars either vertically or horizontally <br>
**Scatter Plot**
- Uses cartesian coordinates to display values for two variables
- Value of one variable determines the position on the horizontal axis, and the value of the other variable determines teh position on the vertical axis
- **Uses**
  - Examine the relationship between two continuous variables
  - Investigate patterns or trends in data
  - Detect outliers or unusual observations
  - Identify clusters or groups in the data
<br>

**Box plot**
- Displays the distribution of a dataset along with key statistical measures
- Represents the interquartile range as a box, a median line and whiskers indicating range
- Represents outliers as individual data points beyond whiskers
<br>

**Histograms**
- Graphical representation of dataset distribution
- Shows frequency or relative frequency with intervals
- Bars represent the data count in each interval

## **Introduction to Matplotlib**
**Matplotlib - history**
- John Hunter(1968-2012)
- Matplotlib was initially developed as an EEG/ECoG visualization libraries in python

**Matplotlib Architecture**
1. Backend Layer (FigureCanvas, Renderer, Event)
2. Artist Layer (Artist)
3. Scripting Layer (pyplot)

**Backend Layer** <br>
Has three built-in abstract interface calsses:
1. **FigureCanvas:** matplotlib.backend_bases.FigureCanvas
    - Encompasses the area onto which the figure is drawn
2. **Renderer:** matplotlib.backend_bases.Renderer
   - Knows how to draw on the FigureCanvas
3. **Event:** matplotlib.backend_bases.Event
   - Handles user inputs such as keyboard strokes and mouse cliks

**Artist Layer** <br>
Comprised of one main object - Artist:
- Knows how to use the Renderer to draw on the canvas.
- Title, lines, tick labels, and images, all correspond to individual Artist instances.
- **Two types of Artist objects:**
    1. **Primitive** Line2D, Rectangle, CIrcle and Text.
    2. **Composite** Axis, Tick, Axes, and Figure.

**Scripting Layer** <br>
It is comprised mainly of Pyplot, a scripting interface lighter tahn the Artist Layer.

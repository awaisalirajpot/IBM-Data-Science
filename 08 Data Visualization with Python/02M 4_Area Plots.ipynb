```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot') # Optional: for ggplot-like style
print('Matplotlib version: ', mpl.__version__)
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

```python
df_can.head(3)
```

```python
df_can.set_index(['Country'], inplace = True)
```

```python
df_can.head(3)
```

```python
years = list(map(str, range(1980, 2014)))
```

### Area Plots

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
```

```python
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'area',
             stacked= False,
            figsize=(20,10))
plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('years')
plt.ylabel('Numbers of Imigrants')
plt.show()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area', figsize=(20,10))
plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_top5.plot(kind='area',
            alpha=0.25,
            stacked=False,
            figsize=(20,10))
plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

#### Two types of Plotting
There are two styles/options of plotting with matplotlib:
1. Plotting using the Artist layer
2. Plotting using the scripting layer

**Option 1**: Scripting Layer (procedural method) <br>
You can add elements by calling different methods of plt like plt.title(..), plt.xlabel(...) etc

```python
df_top5.plot(kind='area', 
            alpha=0.35, figsize=(30,10))
plt.title('Immigration trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

**Option 2**: Artist layer (Object oriented method) <br>
You can use an Axes instance of your current plot and store it in a variable (eg. ax). You can add more elements by calling methods with a little change in suntax (by adding 'set_' to previous methods). For example, use ax.set_title() instead of plt.title() to add title, or as.set_xlabel() instead of plt.xlabel() to add label to the x-axis.

```python
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(40, 10))
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number')
ax.set_xlabel('Years')
plt.show()
```

### Histograms

```python
count, bin_edges = np.histogram(df_can['2013'])
print(count)
print(bin_edges)
```

```python
df_can['2013'].plot(kind='hist', figsize=(8,5))
plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.xlabel('Number of Immigrants')
plt.ylabel('Number of Countries')
plt.show()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
df_can['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)
plt.title('Histogram of Immigration from 195 countries in 2013')
plt.xlabel('Number of Immirati')
plt.ylabel('Number of COuntries')
plt.show()
```

### **Problems**
 

```python
# Problem 1: How many Elementary Schools are in the dataset?
%sql select count(`Elementary, Middle, or High School`) from CHICAGO_PUBLIC_SCHOOLS_DATA where `Elementary, Middle, or High School` = 'ES'
```

```python
#What is the highest Safety Score?
%sql select max(SAFETY_SCORE) as Max_Safety_Score from CHICAGO_PUBLIC_SCHOOLS_DATA
```

```python
#which school has highest safety score
%sql select name_of_school as name_of_school, max(safety_score) as Max_Safety_Score from chicago_public_schools_data
```

```python
#Since we already found the max score so we can also use wher class
%sql select name_of_school from chicago_public_schools_data where safety_score = 99;
```

```python
# What are the top 10 Schools with the highest "Average Student Attendance"
%sql select name_of_school from chicago_public_schools_data where AVERAGE_STUDENT_ATTENDANCE = 98.40
```

```python
%sql select max(average_student_attendance) from chicago_public_schools_data
```

```python
# Correct version
%sql Name_of_School, Average_student_Attendance from chicago_public_schools_data \
order by average_student_attendance desc nulls last limit 10
```

```python
%sql select Name_of_School, Average_Student_Attendance from CHICAGO_PUBLIC_SCHOOLS_DATA \
    order by Average_Student_Attendance desc nulls last limit 10 
```

```python
# Retrieve the list of 5 Schools with the lowest Average Studet Attendance sorted in Asccdning order based on attendace
%sql select name_of_school, average_student_attendance \
from chicago_public_schools_data \
order by average_student_attendance \
limit 5
```

```python
# Now remove the % sign from the above result set

%sql SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '') \
     from CHICAGO_PUBLIC_SCHOOLS_DATA \
     order by Average_Student_Attendance \
     LIMIT 5
```

```python
%sql select name_of_school, replace(average_student_attendance, '%', '') \
from chicago_public_schools_data \
order by average_student_attendance \
limit 5
```

```python
# Which schools have average student attendance lower than 70%
%sql select name_of_school, replace(average_student_attendance, '%', '') \
from chicago_public_schools_data \
where average_student_attendance < 70 \
order by average_student_attendance
```

```python
# Get the total College Enrollment for each Community Area
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
   from CHICAGO_PUBLIC_SCHOOLS_DATA \
   group by Community_Area_Name 
```

```python
# Get the 5 community areas with the least total college enrollments
%sql select community_area_name, sum(college_enrollment) as College_Enrollment \
from chicago_public_schools_data \
group by community_area_name \
order by sum(college_enrollment) \
limit 5
```

```python
# list 5 schools with lowest safety score
%sql select min(safety_score) from chicago_public_schools_data
```

```python
%sql select name_of_school, safety_score \
from chicago_public_schools_data \
where safety_score != 'None' \
order by safety_score \
limit 5
```

```python
%sql select avg(safety_score)\
from chicago_public_schools_data \
where `Elementary, Middle, or High School` = 'MS'
```

```python
%sql select community_area_number, community_area_name, hardship_index from CENSUS_DATA \
   where community_area_number in \
   ( select community_area_number from CHICAGO_PUBLIC_SCHOOLS_DATA order by college_enrollment desc limit 1 )
```

### **Create & Access SQLite database using Python**
**Objective**
- Create a database

*SQLite is a software library taht implements a self-contained serverless, zer0-configuration, transactional SQL database enigne. SQLite is the most widely deployed SQL database engine in the world*

### Task 1: Creaet database using SQLite

```python
#Install and load sqlite3
#!pip install sqlite3
import sqlite3
```

```python
conn = sqlite3.connect('INSTRUCTOR.db')
```

*Cursor class is an instance using which you can invoke methods that execute SQLie statements, fetch data from the results sets of the queiries. You can crete cursor object using the cursoer() method of the conecction object/class.*


```python
#Cursor Object
cursor_obj = conn.cursor()
```

### Task 2: Create a table in the Database

### Task 4: Query Data in the table

```python
statement = """select * from instructor"""
cursor_obj.execute(statement)
print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

```python
# Fetch few rows from the table
statement = "select * from instructor"
cursor_obj.execute(statement)
print("All the data")
output_many = cursor_obj.fetchmany(2)
for row_many in output_many:
    print(row_many)
```

```python
# Fetch only Fname in the table
statement = "select fname from instructor"
cursor_obj.execute(statement)
print("ALL the data")
output_fname = cursor_obj.fetchall()
for fname in output_fname:
    print(fname)
```

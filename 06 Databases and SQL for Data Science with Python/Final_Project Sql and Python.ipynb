### **Store the datasets in database tables**
To analyze the data using SQL, it first needs to be loaded into SQLite Db. We will create three tables in as under:
1. CENSUS_DATA
2. CHICAGO_PUBLIC_SCHOOLS
3. CHICAGO_CRIME_DATA


```python
census.to_sql('CENSUS_DATA', con, if_exists='replace', index=False, method='multi')
school.to_sql('CHICAGO_PUBLIC_SCHOOLS', con, if_exists='replace', index=False, method='multi')
crime.to_sql('CHICAGO_CRIME_DATA', con, if_exists='replace', index=False, method='multi')
```

```python
# Now lets retrieve number of tables in the FinalAssignment
%sql select name as All_Tables from sqlite_master where type = 'table'
```
